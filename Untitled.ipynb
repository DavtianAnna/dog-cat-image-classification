```python
X = []
y = []
for i in os.listdir('animals/'):
    for j in os.listdir(f'animals/{i}/'):
        try:
            image = mpimg.imread(f'animals/{i}/{j}')
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
            image = cv2.resize(image, (512, 512))        
            image = image.reshape(512, 512, 1)
            X.append(image)
            y.append(i)  
        except:
            continue
X = np.array(X, dtype="float32") / 255.0
y = np.array(y)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import  LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
label = LabelEncoder()
```

```python
y_train = label.fit_transform(y_train)  
y_val = label.transform(y_val)
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, num_classes=2)
y_val = to_categorical(y_val, num_classes=2)
```

```python
from keras import Sequential
from keras.optimizers import Adam
from keras.layers import Dropout
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=16, kernel_size=(3, 3), activation='relu', input_shape=(512, 512, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(units=40, activation='relu'))
model.add(Dense(units=30, activation='relu'))
model.add(Dense(units=17, activation='relu'))

model.add(Dense(units=2, activation='softmax'))
model.compile(optimizer=Adam(learning_rate=0.00005), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor = 'val_loss', patience = 3, restore_best_weights = True)
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, batch_size=5, callbacks=[early_stop])
```

```python
plt.plot(model.history.history['loss'], label = 'loss')
plt.plot(model.history.history['val_loss'], label = 'val_loss')

plt.legend()
plt.show()
```

```python
plt.plot(model.history.history['accuracy'], label = 'accuracy')
plt.plot(model.history.history['val_accuracy'], label = 'val_accuracy')

plt.legend()
plt.show()
```
